### Checking and defining the Outliers

# Bivariate Analysis

```python
df.head()
```

### Which country produces maximum sugarcane ?

```python
new_df = df[["Country","Production(Tons)"]].set_index("Country")
```

```python
new_df
```

### Creating new column for the "Production Percentage" in the new Dataframe

```python
new_df["Production(Tons)_percent"] = new_df["Production(Tons)"]*100/new_df["Production(Tons)"].sum()
```

```python
new_df
```

### Creating a Pie Chart for the new Dataframe

```python
new_df["Production(Tons)_percent"].plot(kind = "pie", autopct = "%.2f")
```

Brazil, India and China contribute to almost 65% of production of sugarcane

### Bar Chart for the above Analysis

```python
df[["Country","Production(Tons)"]].set_index("Country").sort_values("Production(Tons)", ascending = False).head(15).plot(kind = "bar")
```

#### Creating a Bar plot with top 15 rows of the original dataframe

```python
ax = sbn.barplot(data = df.head(15),  x= "Country", y = "Production(Tons)")

ax.set_xticklabels(ax.get_xticklabels(),rotation =90)

mat.show()
```

The country "Brazil" produces maximum sugarcane out of all countries .   (Seleo Brazil)

### Which country has highest land?

```python
df_acr = df.sort_values("Acreage(Hectare)", ascending = False).head(15)  # Taking the top 15 values of the original dataset

px = sbn.barplot(data = df_acr,  x= "Country", y = "Acreage(Hectare)")

px.set_xticklabels(ax.get_xticklabels(),rotation =90)

mat.show()
```

"Brazil" has the largest amount of land dedicated for Sugarcane Production .   (Seleo Brazil)

### Which country has highest yield per hectare?

```python
df_yield = df.sort_values("Yield(Kg/Hectare)", ascending = False).head(15)  # Taking the top 15 values of the original dataset

dx = sbn.barplot(data = df_yield,  x= "Country", y = "Yield(Kg/Hectare)")

dx.set_xticklabels(ax.get_xticklabels(),rotation =90)

mat.show()
```

Guatemala has the highest yield(kg/hectare) for sugarcane . All hail the Latin America Countries.

### Which country has highest production?

```python
df_yield = df.sort_values("Production_per_person(Kg)", ascending = False).head(15)  # Taking the top 15 values of the original dataset

hx = sbn.barplot(data = df_yield,  x= "Country", y = "Production_per_person(Kg)")

hx.set_xticklabels(ax.get_xticklabels(),rotation =90)

mat.show()
```

Production per Person is highest in Paraguay . (Told ya, latin American countries are the best in the world)
